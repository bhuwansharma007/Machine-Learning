```python
import numpy as np
import matplotlib.pyplot as ptl

##For plotting a graph we need 2 points x and y, which we treat them as arrays(x and y)
##Scatter is use to scatter them as points on graph

x=[1,2,3]
y=[4,5,6]
ptl.scatter(x,y)
ptl.show()
```

```python
##plot() is use to make a line
x=[1,2,3]
y=[4,5,6]
ptl.plot(x,y)
ptl.show()
```

```python
##Scatter + Plot
x=[1,2,3]
y=[4,5,6]
ptl.plot(x,y)
ptl.scatter(x,y)
ptl.show()
```

```python
#Plotting 2 different lines
x=[1,2,3]
y=[4,5,6]
ptl.plot(x,y)
x=[2,4,6]
y=[3,5,7]
ptl.plot(x,y)
ptl.show()
```

```python
#Playing with 3rd argument of plot
#First character of 3rd argumentis for colour and Second argument for how to show our graph
x=[1,2,3]
y=[4,5,6]
ptl.plot(x,y,"r-.")
ptl.show()
```

```python
#Plotting x^3
x=np.array([1,2,3,4])
y=x**3
ptl.plot(x,y)
ptl.show()
```

```python
#Since curve is not smooth, we change our increment value of graph using np.arange(start,end,increment)
x=np.arange(0,5,0.1)
y=x**3
ptl.plot(x,y)
ptl.show()
```

```python
#If we give only 1 array to plot
##It will take that array as y-axis and assume x-axis in range(0.....size of x-1) i.e, here(0,1,2)
x=[1,2,3]
ptl.plot(x)
ptl.show()
```

```python
############################################### CUSTOMIZING GRAPHS ##################################################

#Adding x-axis, y-axis, and title to graph->
#just before show write->
x=np.arange(0,5,0.1)
y=x**3
y2=x**2
ptl.plot(x,y)
ptl.plot(x,y2)
ptl.xlabel("x")
ptl.ylabel("y")
ptl.title("Matplotlib Demo")
ptl.show()
```

```python
###To add legend we use legend, but while plotting we should pass label also

x=np.arange(0,5,0.1)
y=x**3
y2=x**2
ptl.plot(x,y,label="x^3")
ptl.plot(x,y2,label="x^2")
ptl.xlabel("x")
ptl.ylabel("y")
#By default it won't show label on graph, so we use legend
ptl.legend()
ptl.title("Matplotlib Demo")
ptl.show()
```

```python
#Changing the start index & end index of x-axis and y-axis using axis([Xstart,Xend,Ystart,Yend])
x=np.arange(0,5,0.1)
y=x**3
y2=x**2
ptl.plot(x,y,label="x^3")
ptl.plot(x,y2,label="x^2")
ptl.xlabel("x")
ptl.ylabel("y")
ptl.axis([0,10,0,150])
#By default iw won't show label on graph, so we use legend
ptl.legend()
ptl.title("Matplotlib Demo")
ptl.show()
```

```python
#Making a Grid using grid()
x=np.arange(0,5,0.1)
y=x**3
y2=x**2
ptl.plot(x,y,label="x^3")
ptl.plot(x,y2,label="x^2")
ptl.xlabel("x")
ptl.ylabel("y")
ptl.axis([0,10,0,150])
#By default iw won't show label on graph, so we use legend
ptl.legend()
ptl.title("Matplotlib Demo")
ptl.grid()
ptl.show()
```

```python
#Adding Random Text on Graph using text(Xaxis,Yaxis,"text")
x=np.arange(0,5,0.1)
y=x**3
y2=x**2
ptl.plot(x,y,label="x^3")
ptl.plot(x,y2,label="x^2")
ptl.xlabel("x")
ptl.ylabel("y")
ptl.axis([0,10,0,150])
#By default iw won't show label on graph, so we use legend
ptl.legend()
ptl.title("Matplotlib Demo")
ptl.grid()
ptl.text(4,60,"RR")
ptl.show()
```

```python
################################################## PIE GRAPHS #########################################################

#We divide our pie graph in form of splits, i.e, it can be either 2,3 4... splits, we represent splits in form of array

sizes=[1,2,3,4] #These are part of graph(splits)
ptl.pie(sizes)
ptl.show()
```

```python
sizes=[1,2,3,4] #These are part of graph(splits)
ptl.pie(sizes)
#It is not round, or width & height are not equal so we make them equal->
ptl.axis("equal")
ptl.show()
```

```python
#Changing colors using colors array and passing it to ptl.pie()function->
sizes=[1,2,3,4] #These are part of graph(splits)
colors=["red","blue","magenta","orange"]
ptl.pie(sizes,colors=colors)
#It is not round, or sizes are not equal so we make them equal->
ptl.axis("equal")
ptl.show()
```

```python
#Adding Title to Pie Graph and labels to each split->
sizes=[1,2,3,4] #These are part of graph(splits)
colors=["red","blue","magenta","orange"]
labels=["A","B","C","D"]
ptl.title("PIE GRAPH IN ACTION")
ptl.pie(sizes,colors=colors,labels=labels)
#It is not round, or sizes are not equal so we make them equal->
ptl.axis("equal")
ptl.show()
```

```python
#Adding data inside each split, i.e, percentage split among each split using autopct="%.1%%"(1 is places after decimal)
sizes=[1,2,3,4] #These are part of graph(splits)
colors=["red","blue","magenta","orange"]
labels=["A","B","C","D"]
ptl.title("PIE GRAPH IN ACTION")
ptl.pie(sizes,colors=colors,labels=labels,autopct="%.1f%%") #In order to show % at end we need to add 2 percentages
#It is not round, or sizes are not equal so we make them equal->
ptl.axis("equal")
ptl.show()
```

```python
#Exploding some splits of Pie Graph using explode array and passing it into ptl.pie()
sizes=[1,2,3,4] #These are part of graph(splits)
explode=[0,0,0.2,0.1]
colors=["red","blue","magenta","orange"]
labels=["A","B","C","D"]
ptl.title("PIE GRAPH IN ACTION")
ptl.pie(sizes,colors=colors,labels=labels,autopct="%.2f%%",explode=explode)
ptl.axis("equal")
ptl.show()
```

```python
#By default pie chart split is counterclock, we can change it by changing counterclock to false inside ptl.pie()
sizes=[1,2,3,4] #These are part of graph(splits)
explode=[0,0,0.2,0.1]
colors=["red","blue","magenta","orange"]
labels=["A","B","C","D"]
ptl.title("PIE GRAPH IN ACTION")
ptl.pie(sizes,colors=colors,labels=labels,autopct="%.2f%%",explode=explode,counterclock=False)
ptl.axis("equal")
ptl.show()

# Pie chart splitis clockwise now :)
```

```python
#Also by default pie Chart split start from 0 degree angle, we can change that by giving startangle inside ptl.pie()
sizes=[1,2,3,4] #These are part of graph(splits)
explode=[0,0,0.2,0.1]
colors=["red","blue","magenta","orange"]
labels=["A","B","C","D"]
ptl.title("PIE GRAPH IN ACTION")
ptl.pie(sizes,colors=colors,labels=labels,autopct="%.2f%%",explode=explode,counterclock=False,startangle=100)
ptl.axis("equal")
ptl.show()

###See here A start from 100 degrees and pie chart splitis clockwise also :)
```
