```python
import numpy as np
data=np.loadtxt("data.csv",delimiter=',')
print(data)
data.shape # data is a 2d array which contain only 1 feature(x) and its corresponding output(y)
```

```python
# x and y are 1 D arrays
x=data[:,0]
y=data[:,1]
print(x.shape)
print(y.shape)
print(x)
print(y)
```

```python
#Splitting data into training and testing
from sklearn import model_selection
# 75% for training and 25% for testing, so it is by-default, since no argument given
X_train, X_test, Y_train, Y_test =model_selection.train_test_split(x,y)
```

```python
#Splitted into 4 1d Arrays
X_train, X_test, Y_train, Y_test
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
#We need to fit this X_train and Y-train i.e, train the algorithm

def fit(X_train, Y_train):
    num=(X_train*Y_train).mean()-X_train.mean()*Y_train.mean()
    den=(X_train**2).mean()-X_train.mean()**2
    
    m=num/den
    c=Y_train.mean()-m*X_train.mean()
    
    return m,c
```

```python
#Predection against the training and testing data
def predict(x,m,c):
    
    return m*x+c
```

```python
#Score of our algo i.e, coefficient of determination
def score(y_predicted,y_true):
    u=((y_true-y_predicted)**2).sum()
    v=((y_true-y_true.mean())**2).sum()
   
    return 1-u/v;
```

```python
#Cost of training or testing the algorithm, i.e, how far our m & c(predicted line/output) against the actual value
##Calculating the mean optimized cost of all the data points
def cost(x_train,y_train,m,c):
    
    return ((y_train-(m*x_train+c))**2).mean()
```

```python
m,c=fit(X_train,Y_train)

y_train_predicted=predict(X_train,m,c)
y_test_predicted=predict(X_test,m,c)

training_score=score(y_train_predicted,Y_train)
testing_score=score(y_test_predicted,Y_test)

training_cost=cost(X_train,Y_train,m,c)
testing_cost=cost(X_test,Y_test,m,c)

print("M, C: ",m,c)
print("Score on train: , Score on test: ",training_score,testing_score)
print("Cost on train: , Cost on test: ",training_cost,testing_cost)
```
