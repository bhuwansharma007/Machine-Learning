```python
############################################## HANDLING NANS ########################################################
```

```python
##Making some entries in our dataset as NaNs, using Numpy
df.iloc[2:4,1:3]=np.nan
df
```

```python
##Dropping NaNs
df.dropna(inplace=True)
df
```

```python
##Filling NA entries by the mean of their column
df.sw.fillna(df.sw.mean(), inplace=True)
print(df.head())

df.pl.fillna(df.pl.mean(), inplace=True)
print(df.head())
```

```python
##OR we can fill the NA with the mean od their flower_type

a=df[df.flower_type=="Iris-setosa"]
a.sw.fillna(a.sw.mean(),inplace=True)
print(a.head())

a.pl.fillna(a.pl.mean(),inplace=True)
print(a.head())
```
